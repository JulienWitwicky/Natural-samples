```python
import serial
import csv
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

#ser.close()
# DÃ©finir la connexion sÃ©rie avec l'Arduino
SERIAL_PORT = 'COM6'  # Modifiez le port selon votre configuration
BAUD_RATE = 115200

ser = serial.Serial(SERIAL_PORT, BAUD_RATE)

# Nom du fichier CSV
bag=input('Bag number ?')
filename = 'EGRIP_S1_'+bag

# Fonction pour sauvegarder les donnÃ©es dans un fichier CSV
def save_data():
    with open(filename+'.csv', 'w', newline='') as csvfile:
        writer = csv.writer(csvfile)
        writer.writerow(['Time', 'Conductivity'])
        
        try:
            while True:
                line = ser.readline().decode('utf-8').strip()
                values = line.split(', ')
                if len(values) == 2:  # Assurez-vous qu'il y a bien deux valeurs
                    time_value = float(values[0])
                    conductivity_value = float(values[1])
                    
                    # Ecrire dans le CSV
                    writer.writerow([float(values[0]), float(values[1])])
                    print(f"Data saved: {values[0]}, {values[1]}")
        except KeyboardInterrupt:
            print("Data collection stopped.")
        except Exception as e:
            print(f"Error: {e}")
        finally:
            ser.close()

# ExÃ©cutez la fonction pour sauvegarder les donnÃ©es
save_data()
```

```python
# scan speed in cm/s
speed=2
# core length in cm
length=float(input('Core length in mm ?'))
length=length/10
print(str(length)+' cm')

# import csv file and plot the raw data. Delete the very first data

df=pd.read_csv(filename+'.csv',sep=',')

df=df.loc[(df['Conductivity'].shift(1)-df['Conductivity'])>0]
df.plot(title='Full acquisition', x='Time',y='Conductivity',figsize=[10,5], grid=True)
df=df.loc[df['Time']>=10/speed]

# ice core detection (start and end) and plot of Conductivity VS Length

#t0=10
#tf=400
#core=df.loc[df['Time']>=t0]
#core=core.loc[core['Time']<tf]

cond_limit=300
core=df.loc[df['Conductivity']>cond_limit]

core.plot(title='Acquisition with conductivty threshold', x='Time',y='Conductivity',figsize=[10,5], grid=True)
plt.show()
core['Length']=(core['Time']-core['Time'].iloc[0])*speed

if ((core['Length'].iloc[-1]-core['Length'].iloc[0]<length*1.01) and (core['Length'].iloc[-1]-core['Length'].iloc[0]>length*0.99)):
    fig, ax1 = plt.subplots()
    core.plot(title=filename, x='Length',y='Conductivity',grid=True,figsize=[12,6],color='red',ax=ax1)
    
else:
    t0=float(input('Core start time ?'))
    tf=float(input('Core end time ?'))
    core=df.loc[df['Time']>=t0]
    core=core.loc[core['Time']<tf]
    core['Length']=(core['Time']-core['Time'].iloc[0])*speed
    fig, ax1 = plt.subplots()
    core.plot(title=filename, x='Length',y='Conductivity',grid=True, figsize=[12,6],color='red',ax=ax1)

#save new csv file and plots
core.to_csv("data/"+filename+'_cond.csv',sep=',') 
plt.savefig("data/"+filename+'_cond.pdf', format='pdf')
plt.savefig("data/"+filename+'_cond.png', format='png')
```
